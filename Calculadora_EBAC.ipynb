```python
# Definição das variáveis

valor_1 = 0.0
valor_2 = 0.0
operacao = 0
```

```python
# Consulta do tipo das variáveis

tipo_valor_1 = type(valor_1)
tipo_valor_2 = type(valor_2)
tipo_operacao = type(operacao)

print(tipo_valor_1)
print(tipo_valor_2)
print(tipo_operacao)
```

```python
# Atribuição dos valores coletados do usuário

valor_1 = float(input("Digite o primeiro valor: "))
valor_2 = float(input("Digite o segundo valor: "))
operacao = int(input("Digite a operação desejada: \n 1 - Adição \n 2 - Subtração \n 3 - Multiplicação \n 4 - Divisão \n"))
```

```python
# Tratativa de erro da escolha de operação do usuário

while (operacao != 1) and (operacao != 2) and (operacao != 3) and (operacao != 4):
  print("Digite uma opção de operação válida!")
  operacao = int(input("Digite a operação desejada"))
```

```python
# Resultado da operação com base na escolha da operação

if (operacao == 1):
    print(valor_1 + valor_2)
elif (operacao == 2):
      print(valor_1 - valor_2)
elif (operacao == 3):
      print(valor_1 * valor_2)
elif (operacao == 4):
      if (valor_2 == 0):
        print("Não é possível dividir por zero!")
      else:
        print(valor_1/valor_2)
```
